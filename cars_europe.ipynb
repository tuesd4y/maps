```python
import geopandas as gpd
import shapely
import geoplot as gplt
import geoplot.crs as gcrs
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
# read in country shapes
countries = gpd.read_file('cars/CNTR_RG_60M_2020_4326.geojson')
country_population = pd.read_csv('cars/API_SP.POP.TOTL_DS2_en_csv_v2_1637443.csv', header=2, index_col=1)
country_population.head()
# countries.head()
```

```python
# df_years = for year in range(0, 2020-1990)] = 
cars = pd.read_excel('cars/1990â€“2019_PC_by_country_EU+EFTA.xlsx', sheet_name=None, header = 2, index_col=0)

# cars['1990'].transpose()
for key in cars:
    cars[key] = cars[key] 
cars['1990']
```

```python
box = shapely.geometry.LineString(coordinates = [[-25.312500,34.379713],[41.484375,71.272595]])

box = gpd.GeoSeries([box])
box = box.set_crs(epsg=4326)
box = box.envelope[0]

countries = countries[countries.intersects(box)]


data = []
# assign car numbers to countries
for y in range(0, 30):
    year = 1990 + y
    c = cars[f'%d' % year]
    pop = country_population[[f'%d' % year]]
    for month in range(0, 12):
        df = countries[['NAME_ENGL', 'geometry', 'ISO3_CODE']]
        c_month = c.iloc[:,month]
        
        def get_car_count(d):
            try: 
                return c_month[d['NAME_ENGL']]
            except: 
                return np.NaN
                
        df['cars'] = df.apply(get_car_count, axis = 1)
        df['carspp'] = df.dropna().apply(lambda r: r['cars'] / pop.iloc[:,0][r['ISO3_CODE']] * 1000, axis=1)
    
        data.append(df)
        
        
        
# cars['1990'].iloc[:,0]['Austria']
data[0].plot(column = 'cars')


pop = country_population[[f'%d' % year]]
pop
```

```python
# plot
import geoplot as gplt
import matplotlib as mpl


cmap = sns.color_palette("Reds", as_cmap=True)
# cmap = sns.color_palette("YlOrBr", as_cmap=True)
# norm = mpl.colors.Normalize(vmin=0, vmax=14)
norm = mpl.colors.Normalize(vmin=0, vmax=210)


# norm = mpl.colors.LogNorm(vmin=1, vmax = 120)
cbar = plt.cm.ScalarMappable(norm=norm, cmap=cmap)

def plot(data: gpd.GeoDataFrame, i: int):
    year = 1990+i/12
    month = 1+i%12
    fig, ax = plt.subplots(1, 1, figsize=(16, 12))
    gplt.choropleth(data, hue = 'carspp', cmap = cmap, ax=ax, 
                     extent = (-13, 35, 30, 70), norm = norm,
                     legend = False, linewidth= .1)
    plt.suptitle(f'Car registrations per country, %02d/%04d' % (month, year), fontsize=16)
    plt.subplots_adjust(top=0.94)
    plt.colorbar(cbar, ax = ax, fraction = .04, orientation = 'horizontal', pad = .05, 
                 label = 'new registrations per 1000 persons')
    fig.savefig(f"frames/frame_%04d.png" % i, 
                dpi=100, bbox_inches='tight')
    plt.close()
    
# for i in range(0, (2019-1990+1)*12):
#     plot(data[i], i)
# np.max([d['carspp'].max() for d in data])
# plot(data[11], 1990, 11)

def plot_avg(i: int):
    year = i + 1990
    
    # sum up values for the whole year into d
    d= data[i*12]
    for m in range(1,12):
        d['carspp'] = d['carspp'].add(data[i*12+m]['carspp'], fill_value=0)
        
    fig, ax = plt.subplots(1, 1, figsize=(32, 24))
    gplt.choropleth(d, hue = 'carspp', cmap = cmap, ax=ax, 
                     extent = (-13, 35, 30, 70), norm = norm,
                     legend = False, linewidth= .05)
    plt.suptitle(f'Car registrations per country, {year}', fontsize=36)
    plt.subplots_adjust(top=0.94)
    plt.colorbar(cbar, ax = ax, fraction = .04, orientation = 'horizontal', pad = .05) \
            .set_label(label='new registrations per 1000 persons', size=15)
    fig.savefig(f"frames/frame_%02d.png" % i, 
                dpi=100, bbox_inches='tight')
    plt.close()
    print('', d['carspp'].min(),  d['carspp'].max())


        
        
cmap = sns.color_palette("Reds", as_cmap=True)
norm = mpl.colors.Normalize(vmin=0, vmax=120)
cbar = plt.cm.ScalarMappable(norm=norm, cmap=cmap)

for i in range(0, 2019-1990+1):
    plot_avg(i)

# data[0].head()
```
